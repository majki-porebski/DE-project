```python
import torch
from torch import nn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons
from torchmetrics import Accuracy

```

```python
#generate dataset
RANDOM=6  
X, y = make_moons(n_samples = 1000, shuffle=True, noise=0.4, random_state=RANDOM)
```

```python
# create dataset with X1, X2, y columns 
moon = pd.DataFrame({"X1": X[:,0],
                    "X2": X[:,1],
                    "y": y})
```

```python
plt.scatter(x=X[:, 0], 
            y=X[:, 1], 
            c=y, 
            cmap=plt.cm.RdYlBu);
```

```python
#set device agnostic code

device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
#turn data into tensors

X = torch.from_numpy(X).type(torch.float)
y = torch.from_numpy(y).type(torch.float)
```

```python
#test split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=RANDOM)
len(train_X), len(test_X), len(train_y), len(test_y)
```

# 2. Build a model
* forward pass
* return
* pick optimazer and loss function
* create a loop

```python
class MoonClassifierV0(nn.Module):
  def __init__(self, input_size, output_size,hidden_size, random_state=6):
    super().__init__()
    ## define layers
    self.linear1 = nn.Linear(in_features=input_size, out_features=hidden_size)
    self.linear2 = nn.Linear(in_features=hidden_size, out_features=hidden_size)
    self.linear3 = nn.Linear(in_features=hidden_size, out_features=hidden_size)
    self.linear4 = nn.Linear(in_features=hidden_size, out_features=output_size)
    self.relu = nn.ReLU()
    self.tanh = nn.Tanh()

  ### forward
  def forward(self, x):
    return self.linear4(self.tanh(self.linear3(self.relu(self.linear2(self.relu(self.linear1(x)))))))

model_0 = MoonClassifierV0(input_size=2, hidden_size=512, output_size=1)
model_0 = model_0.to(device)
```

```python
# choose loss function and optimizer
loss_fn = nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(params=model_0.parameters(),
                            lr=0.1)
```

```python
# predicted values
#y_logits = model_0(train_X).squeeze()
#y_logits.shape, train_y.shape
acc_fn = Accuracy().to(device) # send accuracy function to device
acc_fn
```

```python
torch.manual_seed(6)
epochs = 1001

# Put data to target device
train_X, train_y = train_X.to(device), train_y.to(device)
test_X, test_y = test_X.to(device), test_y.to(device)

#training loop
for epoch in range(epochs):
  model_0.train()
  
    # 1. Forward pass
  y_logits = model_0(train_X).squeeze()
  y_pred = torch.round(torch.sigmoid(y_logits))
  
    # 2. Calculate loss and accuracy
  loss = loss_fn(y_logits, train_y)
  acc = acc_fn(y_pred, train_y.int())
  
    # 3. Zero grad
  optimizer.zero_grad()

    # 4. Backpropagation
  loss.backward()

    # 5. Optimizer step
  optimizer.step()


# testing loop
  model_0.eval()
  with torch.inference_mode():
      # 1. Forward pass
    test_logits = model_0(test_X).squeeze()
    test_pred = torch.round(torch.sigmoid(test_logits))
  
      # 2. Calculate loss and accuracy
    test_loss = loss_fn(y_logits, train_y)
    test_acc = acc_fn(test_pred, test_y.int())

  # Print what's happening
  if epoch % 100 == 0:
    print(f"Epoch: {epoch} | Loss: {loss:.5f}, Acc: {acc:.2f} | TestLoss: {test_loss:.5f}, TestAcc: {test_acc:.2f}")
```

```python


def plot_decision_boundary(model_0, X, y):
  
    # Put everything to CPU (works better with NumPy + Matplotlib)
    model_0.to("cpu")
    X, y = X.to("cpu"), y.to("cpu")

    # Source - https://madewithml.com/courses/foundations/neural-networks/ 
    # (with modifications)
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 101), 
                         np.linspace(y_min, y_max, 101))

    # Make features
    X_to_pred_on = torch.from_numpy(np.column_stack((xx.ravel(), yy.ravel()))).float()

    # Make predictions
    model_0.eval()
    with torch.inference_mode():
        y_logits = model_0(X_to_pred_on)

    # Test for multi-class or binary and adjust logits to prediction labels
    if len(torch.unique(y)) > 2:
        y_pred = torch.softmax(y_logits, dim=1).argmax(dim=1) # mutli-class
    else: 
        y_pred = torch.round(torch.sigmoid(y_logits)) # binary
    
    # Reshape preds and plot
    y_pred = y_pred.reshape(xx.shape).detach().numpy()
    plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
```

```python
# Plot decision boundaries for training and test sets
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Train")
plot_decision_boundary(model_0, train_X, train_y)
plt.subplot(1, 2, 2)
plt.title("Test")
plot_decision_boundary(model_0, test_X, test_y)
```
